## 3. Feature Analysis

### 3.1 Original Features (27 behavioral features)

```python
# Define feature groups (27 original features including AI Assistant)
feature_groups = {
    'Active/Reflective': [
        'activeModeRatio', 'questionsGenerated', 'debatesParticipated',
        'reflectiveModeRatio', 'reflectionsWritten', 'journalEntries'
    ],
    'AI Assistant': [
        'aiAskModeRatio', 'aiResearchModeRatio', 'aiTextToDocsRatio'
    ],
    'Sensing/Intuitive': [
        'sensingModeRatio', 'simulationsCompleted', 'challengesCompleted',
        'intuitiveModeRatio', 'conceptsExplored', 'patternsDiscovered'
    ],
    'Visual/Verbal': [
        'visualModeRatio', 'diagramsViewed', 'wireframesExplored',
        'verbalModeRatio', 'textRead', 'summariesCreated'
    ],
    'Sequential/Global': [
        'sequentialModeRatio', 'stepsCompleted', 'linearNavigation',
        'globalModeRatio', 'overviewsViewed', 'navigationJumps'
    ]
}

all_features = [f for group in feature_groups.values() for f in group]
label_cols = ['activeReflective', 'sensingIntuitive', 'visualVerbal', 'sequentialGlobal']

print(f"üìä Original Features: {len(all_features)}")
print(f"üéØ Target Labels: {len(label_cols)}")
print(f"\nüìã Feature Groups:")
for group, features in feature_groups.items():
    print(f"  {group}: {len(features)} features")
    for feat in features:
        print(f"    - {feat}")
```

### 3.2 Engineered Features (19 additional features)

The system creates 19 engineered features for better performance:

```python
print("üîß Engineered Features (19 total):")
print("\n1Ô∏è‚É£ Ratio Features (4):")
print("   - active_reflective_ratio")
print("   - sensing_intuitive_ratio")
print("   - visual_verbal_ratio")
print("   - sequential_global_ratio")

print("\n2Ô∏è‚É£ Intensity Features (8):")
print("   - active_intensity (questions + debates)")
print("   - reflective_intensity (reflections + journal)")
print("   - sensing_intensity (simulations + challenges)")
print("   - intuitive_intensity (concepts + patterns)")
print("   - visual_intensity (diagrams + wireframes)")
print("   - verbal_intensity (text + summaries)")
print("   - sequential_intensity (steps + linear nav)")
print("   - global_intensity (overviews + jumps)")

print("\n3Ô∏è‚É£ Squared Features (4):")
print("   - activeModeRatio_squared")
print("   - sensingModeRatio_squared")
print("   - visualModeRatio_squared")
print("   - sequentialModeRatio_squared")

print("\n4Ô∏è‚É£ AI Interaction Features (3):")
print("   - ai_active_interaction")
print("   - ai_reflective_interaction")
print("   - ai_sensing_interaction")

print("\nüìä Total Features: 27 original + 19 engineered = 46 features")
```

## 4. Target Label Distribution

```python
# Visualize label distributions
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('FSLSM Dimension Distributions (Current System)', fontsize=16, fontweight='bold')

dimensions = [
    ('activeReflective', 'Active/Reflective', axes[0, 0]),
    ('sensingIntuitive', 'Sensing/Intuitive', axes[0, 1]),
    ('visualVerbal', 'Visual/Verbal', axes[1, 0]),
    ('sequentialGlobal', 'Sequential/Global', axes[1, 1])
]

for col, title, ax in dimensions:
    ax.hist(df[col], bins=23, edgecolor='black', alpha=0.7, color='steelblue')
    ax.set_title(f'{title}\nMean: {df[col].mean():.2f}, Std: {df[col].std():.2f}', fontweight='bold')
    ax.set_xlabel('Score (-11 to +11)')
    ax.set_ylabel('Frequency')
    ax.axvline(df[col].mean(), color='red', linestyle='--', linewidth=2, label='Mean')
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nüìä Label Statistics:")
print(df[label_cols].describe())
```

## 5. Feature Correlations with Target Labels

```python
# Calculate correlations between features and labels
correlations = {}
for label in label_cols:
    corr = df[all_features + [label]].corr()[label].drop(label).sort_values(ascending=False)
    correlations[label] = corr

# Visualize top correlations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Top Feature Correlations with FSLSM Dimensions', fontsize=16, fontweight='bold')

for idx, (label, ax) in enumerate(zip(label_cols, axes.flatten())):
    top_corr = correlations[label].head(10)
    colors = ['green' if x > 0 else 'red' for x in top_corr.values]
    top_corr.plot(kind='barh', ax=ax, color=colors, alpha=0.7)
    ax.set_title(f'{label}\nTop 10 Correlated Features', fontweight='bold')
    ax.set_xlabel('Correlation Coefficient')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Load Trained Models and Evaluate

```python
# Load trained models
models_dir = Path('../models')

print("üì¶ Loading trained models...")
models = {}
scalers = {}

for dimension in label_cols:
    model_path = models_dir / f'{dimension}_model.pkl'
    scaler_path = models_dir / f'{dimension}_scaler.pkl'
    
    if model_path.exists() and scaler_path.exists():
        models[dimension] = joblib.load(model_path)
        scalers[dimension] = joblib.load(scaler_path)
        print(f"  ‚úÖ Loaded {dimension} model")
    else:
        print(f"  ‚ö†Ô∏è  {dimension} model not found")

if len(models) == 4:
    print("\n‚úÖ All 4 models loaded successfully!")
else:
    print(f"\n‚ö†Ô∏è  Only {len(models)}/4 models loaded. Run training script first.")
```

## 7. Model Performance Analysis

```python
# Function to engineer features (same as training script)
def engineer_features(X, feature_cols):
    df_features = pd.DataFrame(X, columns=feature_cols)
    
    # Ratio features
    df_features['active_reflective_ratio'] = df_features['activeModeRatio'] / (df_features['reflectiveModeRatio'] + 0.001)
    df_features['sensing_intuitive_ratio'] = df_features['sensingModeRatio'] / (df_features['intuitiveModeRatio'] + 0.001)
    df_features['visual_verbal_ratio'] = df_features['visualModeRatio'] / (df_features['verbalModeRatio'] + 0.001)
    df_features['sequential_global_ratio'] = df_features['sequentialModeRatio'] / (df_features['globalModeRatio'] + 0.001)
    
    # Intensity features
    df_features['active_intensity'] = df_features['questionsGenerated'] + df_features['debatesParticipated']
    df_features['reflective_intensity'] = df_features['reflectionsWritten'] + df_features['journalEntries']
    df_features['sensing_intensity'] = df_features['simulationsCompleted'] + df_features['challengesCompleted']
    df_features['intuitive_intensity'] = df_features['conceptsExplored'] + df_features['patternsDiscovered']
    df_features['visual_intensity'] = df_features['diagramsViewed'] + df_features['wireframesExplored']
    df_features['verbal_intensity'] = df_features['textRead'] + df_features['summariesCreated']
    df_features['sequential_intensity'] = df_features['stepsCompleted'] + df_features['linearNavigation']
    df_features['global_intensity'] = df_features['overviewsViewed'] + df_features['navigationJumps']
    
    # Squared features
    for col in ['activeModeRatio', 'sensingModeRatio', 'visualModeRatio', 'sequentialModeRatio']:
        df_features[f'{col}_squared'] = df_features[col] ** 2
    
    # AI interaction features
    if 'aiAskModeRatio' in df_features.columns:
        df_features['ai_active_interaction'] = df_features['aiAskModeRatio'] * df_features['activeModeRatio']
        df_features['ai_reflective_interaction'] = df_features['aiResearchModeRatio'] * df_features['reflectiveModeRatio']
        df_features['ai_sensing_interaction'] = df_features['aiTextToDocsRatio'] * df_features['sensingModeRatio']
    
    return df_features.values

# Prepare data
X = df[all_features].values
X_engineered = engineer_features(X, all_features)

# Evaluate each model
if len(models) == 4:
    print("\nüìä Model Performance on Full Dataset:")
    print("=" * 70)
    
    results = {}
    
    for dimension in label_cols:
        y_true = df[dimension].values
        
        # Scale and predict
        X_scaled = scalers[dimension].transform(X_engineered)
        y_pred = models[dimension].predict(X_scaled)
        
        # Calculate metrics
        mae = mean_absolute_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        
        results[dimension] = {'MAE': mae, 'R¬≤': r2, 'RMSE': rmse}
        
        print(f"\n{dimension}:")
        print(f"  MAE:  {mae:.3f}")
        print(f"  R¬≤:   {r2:.3f} ({r2*100:.1f}% accuracy)")
        print(f"  RMSE: {rmse:.3f}")
    
    # Average performance
    avg_r2 = np.mean([r['R¬≤'] for r in results.values()])
    avg_mae = np.mean([r['MAE'] for r in results.values()])
    
    print("\n" + "=" * 70)
    print(f"üìä AVERAGE PERFORMANCE:")
    print(f"  Average R¬≤:  {avg_r2:.3f} ({avg_r2*100:.1f}% accuracy)")
    print(f"  Average MAE: {avg_mae:.3f}")
    print("=" * 70)
```

## 8. Prediction vs Actual Visualization

```python
if len(models) == 4:
    fig, axes = plt.subplots(2, 2, figsize=(14, 12))
    fig.suptitle('Predicted vs Actual Values (Current System)', fontsize=16, fontweight='bold')
    
    for idx, (dimension, ax) in enumerate(zip(label_cols, axes.flatten())):
        y_true = df[dimension].values
        X_scaled = scalers[dimension].transform(X_engineered)
        y_pred = models[dimension].predict(X_scaled)
        
        # Scatter plot
        ax.scatter(y_true, y_pred, alpha=0.3, s=20)
        
        # Perfect prediction line
        min_val = min(y_true.min(), y_pred.min())
        max_val = max(y_true.max(), y_pred.max())
        ax.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect Prediction')
        
        r2 = results[dimension]['R¬≤']
        mae = results[dimension]['MAE']
        
        ax.set_title(f'{dimension}\nR¬≤ = {r2:.3f} ({r2*100:.1f}%), MAE = {mae:.3f}', fontweight='bold')
        ax.set_xlabel('Actual Value')
        ax.set_ylabel('Predicted Value')
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```

## 9. Feature Importance Analysis

```python
if len(models) == 4:
    # Get feature names after engineering
    df_temp = pd.DataFrame(X, columns=all_features)
    X_temp_eng = engineer_features(X[:1], all_features)
    
    # Recreate feature names
    engineered_feature_names = all_features + [
        'active_reflective_ratio', 'sensing_intuitive_ratio', 'visual_verbal_ratio', 'sequential_global_ratio',
        'active_intensity', 'reflective_intensity', 'sensing_intensity', 'intuitive_intensity',
        'visual_intensity', 'verbal_intensity', 'sequential_intensity', 'global_intensity',
        'activeModeRatio_squared', 'sensingModeRatio_squared', 'visualModeRatio_squared', 'sequentialModeRatio_squared',
        'ai_active_interaction', 'ai_reflective_interaction', 'ai_sensing_interaction'
    ]
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 14))
    fig.suptitle('Top 15 Feature Importances by Dimension', fontsize=16, fontweight='bold')
    
    for dimension, ax in zip(label_cols, axes.flatten()):
        # Get feature importances
        importances = models[dimension].feature_importances_
        feature_importance_df = pd.DataFrame({
            'feature': engineered_feature_names,
            'importance': importances
        }).sort_values('importance', ascending=False).head(15)
        
        # Plot
        feature_importance_df.plot(x='feature', y='importance', kind='barh', ax=ax, legend=False, color='steelblue')
        ax.set_title(f'{dimension}', fontweight='bold')
        ax.set_xlabel('Importance')
        ax.set_ylabel('')
        ax.invert_yaxis()
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```
